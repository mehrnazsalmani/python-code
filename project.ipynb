```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

# A , B , C

```python
x = np.random.normal(0,1,100)
a = np.array(x)
eps = np.random.normal(0,0.25,100)
eps = np.array(eps)
y = -1 + 0.5*x + eps
y = np.array(y)
print (y)
```

# C

```python
plt.scatter(x,y, color='r')
plt.title('points',fontname='arial', fontsize=20)
plt.xlabel('X',color='g',fontsize=15)
plt.ylabel('Y',color='g',fontsize=15)
plt.show()
```

# E

```python
from sklearn.model_selection import train_test_split
X = x
Y = y
X_train , X_test , Y_train , Y_test = train_test_split (X , Y , test_size = 0.2 , random_state = 1 )
print( 'x_train : ' , X_train.shape , 'y_train : ', Y_train.shape)
X_train
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error
from sklearn.linear_model import LinearRegression
X_train = X_train.reshape(-1,1)
Y_train = Y_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
lnr = LinearRegression()
lnr.fit(X_train , Y_train)
print(lnr.coef_ ,lnr.intercept_)
```

# F

```python
Y_pred = lnr.predict(X_test)
accuracy = lnr.score(X_test , Y_test)
print(accuracy)
R2 = r2_score(X_test,Y_pred)
print(R2)
plt.scatter(X_test,Y_test)
plt.plot(X_test,Y_pred)
plt.show()
```

```python
plt.scatter(x,y, color='r')
plt.title('points',fontname='arial', fontsize=20)
plt.xlabel('X',color='g',fontsize=15)
plt.ylabel('Y',color='g',fontsize=15)
plt.plot(X_test,Y_pred)
plt.legend(['x,y_pred','x,y'])
plt.show()
```
